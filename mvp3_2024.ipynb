# 1. Objetivo

Considerando a grande diversidade econômica presente em nosso país e tendo grande parte da minha vida acadêmica sido executada em escolas públicas, vivenciei diversas experiências sobre as dificuldades de alguns amigos em participar e permanecer nas instituições de ensino por muitos fatores distintos: falta de recurso para compra de material escolar, alimentação precária, dificuldade com locomoção, falta de infraestrutura familiar, etc.

Diante disso, esse MVP visa relacionar de forma simples o desempenho escolar dos alunos que participaram do ENEM com alguns fatores socioeconômicos da família, sem que para isso sejam utilizados recursos de análise estatística como modelos preditivos, por exemplo.

Para tentar cumprir esse objetivo, serão utilizados dados disponíveis na plataforma aberta "gov.br" para responder as seguintes perguntas:

  1. A escolaridade dos pais influencia as notas dos alunos?

  2. Quais UF´s apresentam maior disparidade entre desempenho escolar e nível socioeconômico?

  3. A infraestrutura familiar (ex.: acesso à internet, telefones, etc) tem impacto significativo no desempenho dos alunos?

  4. A renda familiar impacta diretamente no sucesso acadêmico dos alunos?

## 5.2 Análises

As consultas criadas no item 5.1 buscam responder às questões colocadas no início deste MVP e compõem o objetivo deste projeto. O conjunto de dados inicialmente deveria incluir algumas bases, mas a complexidade de uma análise mais profunda exigiria o desenvolvimento de algoritmos que fugiriam um pouco do objetivo deste trabalho.

Diante do exposto, a análise foi simplificada e direcionada apenas para a base do Enem de 2023, que já incorpora um questionário socioeconômico e permite extrair algumas visões.

Abaixo, seguem as questões a serem tratadas:

1. **A escolaridade dos pais influencia as notas dos alunos?**

   **Resposta:** O resultado de duas consultas que geram a média das notas de Matemática e Linguagens e Códigos indica que sim. A pesquisa separa a escolaridade de pais e mães, mas o que vemos é que as médias das notas são maiores quando os pais possuem maior escolaridade. Filhos de pais que completaram uma pós-graduação apresentaram notas superiores entre 30% e 40% em comparação aos filhos de pais que nunca estudaram. Para uma resposta baseada apenas em números, SIM, é possível afirmar que a escolaridade dos pais influencia nas notas dos alunos.

---

2. **Quais UFs apresentam maior disparidade entre desempenho escolar e nível socioeconômico?**

   **Resposta:** O resultado apresenta o Rio Grande do Sul, São Paulo, Minas Gerais e Distrito Federal figurando nas primeiras posições da lista com praticamente a mesma média de notas. Nos últimos lugares figuram os estados do Maranhão, Ceará, Amazonas e Roraima, na ordem do último para o primeiro. Se considerarmos outros dados, como o Índice de Desenvolvimento Humano (IDH), podemos entender que os dados revelam a veracidade sobre quanto melhor a condição de um local avaliado por sua renda, saúde e educação, melhores as médias de notas dos alunos. E isso se aplica também à parte final da lista quando falamos do contrário. Há certa “tendência” no ranking do IDHM do Brasil. Os municípios que apresentam os melhores IDHMs concentram-se especialmente nas regiões Sudeste e Sul do país. Já os municípios com os piores índices encontram-se especialmente nas regiões Nordeste e Norte. Isso se deve principalmente ao desenvolvimento de cada região.

---

3. **A infraestrutura familiar (ex.: acesso à internet, telefones, etc.) tem impacto no desempenho dos alunos?**

   **Resposta:** O resultado indica que as médias das notas melhoraram com o acesso a computadores com internet. A variação ocorre entre 15% e 20% comparando as médias de quem possui esses recursos com as médias daqueles que não possuem. No entanto, é importante notar que a mera presença da tecnologia não garante resultados positivos, sendo crucial a forma como ela é integrada ao processo de ensino-aprendizagem.

   Há diversos estudos que relacionam os recursos computacionais com melhores resultados no aprendizado. Segue abaixo uma das referências indicadas pelos mecanismos de busca:

   **Uso de Computadores e Resultados Educacionais**
   - **Título:** "The Effect of Home Computer Use on Children’s Cognitive and Non-Cognitive Skills"
   - **Autor:** Fiorini, M.
   - **Ano:** 2019
   - **Resumo:** Este estudo investiga o impacto do uso de computadores em casa nas habilidades cognitivas e não cognitivas das crianças. Os resultados indicam que o uso moderado de computadores está associado a melhores resultados educacionais.
   - **Referência:** [Economics of Education Review](https://www.sciencedirect.com/science/article/abs/pii/S0272775709000931)

---

4. **A renda familiar impacta diretamente no sucesso acadêmico dos alunos?**

   **Resposta:** A consulta feita ao banco de dados do Enem 2023 relata que a renda familiar impacta nas notas, mas é preciso cautela para afirmar sobre sucesso acadêmico. Não há dúvidas de que uma renda familiar melhor permite acesso a melhores recursos para uso acadêmico, mas é necessário enriquecer com mais fontes de dados, como o censo.
